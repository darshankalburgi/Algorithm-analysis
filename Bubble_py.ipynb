```python
import random
import time
import copy

def generate_random_numbers(n):
 return [random.randint(10, 1000) for _ in range(n)]

def measure_sort_time (numbers):
 start_time = time.time()
 numbers.sort ()
 end_time = time.time()
 return end_time - start_time

def jumble_numbers (numbers):
 jumbled = numbers[:]
 random. shuffle (jumbled)
 return jumbled

N= 5
Input_Sizes = generate_random_numbers (N)
Input_Sizes.sort ()

Input_Arrays = []
Sorted_Time_Consumed = []
Rev_Time_Consumed = []
Jumbled_Time_Consumed = []

for size in Input_Sizes:
  Input_Arrays.append(generate_random_numbers(size))

# Process each list in Input_Arrays
for arr in Input_Arrays:
  # Case 1: Sorted input
  Sorted_Input = sorted(arr)
  sort_time = measure_sort_time(copy.deepcopy (Sorted_Input ))
  Sorted_Time_Consumed.append(sort_time)

  # Case 2: Reverse sorted input
  Rev_Sorted_Input = sorted(arr, reverse=True)
  rev_sort_time = measure_sort_time(copy.deepcopy (Rev_Sorted_Input))
  Rev_Time_Consumed. append (rev_sort_time)

  # Case 3: Jumbled input
  Jumbled_Input = jumble_numbers (arr)
  jumbled_sort_time = measure_sort_time(copy.deepcopy (Jumbled_Input ))
  Jumbled_Time_Consumed. append (jumbled_sort_time)

# Display the lists

  print ("Input_Sizes:", Input_Sizes, "Sorted Input")
  for size, sorted_time in zip(Input_Sizes, Sorted_Time_Consumed):
    print (f"{size}-{sorted_time:.6f}")

  print ("Input_Sizes:", Input_Sizes, "Reverse Sorted Input")
  for size, sorted_time in zip(Input_Sizes, Rev_Time_Consumed):
    print (f"{size}-{sorted_time:.6f}")

  print ("Input_Sizes:", Input_Sizes, "Jumbled Input")
  for size, sorted_time in zip(Input_Sizes, Jumbled_Time_Consumed):
    print (f"{size}-{sorted_time:.6f}")
```

```python
import matplotlib.pyplot as plt

def plot_graph(input_sizes, sorted_times, rev_times, jumbled_times):
  plt.figure(figsize=(10, 8))
  plt.plot(input_sizes, sorted_times, label='Sorted Input', marker='o')
  plt.plot(input_sizes, rev_times, label='Reverse Sorted Input', marker='o')
  plt.plot(input_sizes, jumbled_times, label='Jumbled Input', marker='o')
  plt.xlabel('Input Size')
  plt.ylabel ('Time Consumed (seconds) ')
  plt.title('Comparison of Sort Times for Different InputOrders')
  plt.legend ()
  plt.grid(True)
  plt.show()
```

```python
plot_graph(Input_Sizes, Sorted_Time_Consumed, Rev_Time_Consumed,
Jumbled_Time_Consumed)
```
